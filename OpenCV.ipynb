```python
#import word imports pre_existing library
#cv2 is second version, first version is outdated
import cv2
import numpy as np
from google.colab.patches import cv2_imshow #directly imports cv2 and numpy
```

```python
#numpy is a library used to generate an array of number
#black colour:0
#white colour:255
np.zeros(1) #.denotes that this is floating value
```

```python
np.zeros(7) #prints 7 zeros
```

```python
np.zeros(7,dtype='int') #prints 7 zeros in int type
```

```python
np.zeros([100,200])
```

```python
cv2_imshow(np.zeros([100,400])) #100 pixels vertically 400 pixels horizontally, prints a black block/image
```

```python
np.ones([100,400])
```

```python
cv2_imshow(np.ones([100,400]))
# still black, use paint reference, image is still black just got little lighter
# we need 255 to get white image
```

```python
cv2_imshow(np.ones([100,400])*255) #multiply with 255 to get white image
```

```python
# Paint reference: more towards 255 gives lighter shade, more towards 0 gives darker shade
cv2_imshow(np.ones([100,400])*127)
```

```python
np.ones([100,400,3])*127
```

```python
cv2_imshow(np.ones([100,400,3])*255)
```

```python
a=np.ones([100,400,3])*255
a[:,:]=[0,0,0] #B G R
cv2_imshow(a)
#colon indicates from start till end
#before the comma:rows, after comma:column
```

```python
a=np.ones([100,400,3])*255
a[:,:]=[255,255,255] #B G R
cv2_imshow(a)
```

```python
#to print blue colour
a=np.ones([100,400,3])*255
a[:,:]=[255,0,0] #B G R
cv2_imshow(a)
```

```python
#to print green colour
a=np.ones([100,400,3])*255
a[:,:]=[0,255,0] #B G R
cv2_imshow(a)
```

```python
#to print red colour
a=np.ones([100,400,3])*255
a[:,:]=[0,0,255] #B G R
cv2_imshow(a)
```

```python
#to print random colour
a=np.ones([100,400,3])*255
a[:,:]=[255,170,237] #B G R
cv2_imshow(a)
```

```python
#to print green colour
a=np.ones([100,400,3])*255
a[0:100,0:200]=[255,170,237] #B G R
a[0:100,200:400]=[0,0,255]
cv2_imshow(a)
```

```python
#to print green colour
a=np.ones([100,400,3])
a[0:100,0:300]=[255,0,0] #B G R
a[0:50,200:400]=[255,253,116]
cv2_imshow(a)
```
